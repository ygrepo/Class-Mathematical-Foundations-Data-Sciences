```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from math import sqrt
```

```python
def f(x):
    return np.cos(np.pi/2 * x)
vf = np.vectorize(f)
```

```python
def taylor_approximation(x):
    t1 = 1
    t2 = -np.pi**2/8 * (x**2)
    return t1 + t2
vf2 = np.vectorize(taylor_approximation)
```

```python
def projection(x):
    return 0.9 * sqrt(2)/2.0 -0.43 * (sqrt(45/8)) * (x**2-1/3.)
vf3 = np.vectorize(projection)
```

```python
x = np.linspace(-1,1, num=30, endpoint=True)
y1 = vf(x)
y2 = vf2(x)
y3 = vf3(x)
```

```python
fig, ax = plt.subplots(1,1, figsize=(10, 6))
start, end = min(y2), max(y2)  + 0.1
ax.yaxis.set_ticks(np.arange(start, end, 0.1))
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%1.1f'))
ax.set_aspect(1)
ax.set_title("Function approximation", fontsize=10)
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.plot(x,y1,'.k', color="r", label=r"$f(x)=\cos(\frac{\pi}{2}\; x)$")
ax.plot(x,y3, '.', color="g", label=r"$P_{P_2} f$")
ax.plot(x,y2,'.k', color="b", label=r"$T_2 f$")
ax.legend(loc="upper left")
plt.tight_layout()
plt.show();
```
